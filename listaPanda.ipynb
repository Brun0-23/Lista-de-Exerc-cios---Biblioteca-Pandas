```python
##1-Exibir a quantidade de linhas e colunas do data frame.
import pandas as pd
df = pd.read_csv('vendas_de_ingressos.csv')
df.shape
```

```python
'''
4-Renomear as colunas de forma permanente:
Age = Idade
Ticket_Price = Preco_do_Ingresso
Movie_Genre = Genero
Seat_Type = Tipo_de_Assento
Number_of_Person = Numero_de_Pessoas
Purchase_Again = Compraria_Novamente
'''
df.rename(columns = {
    'Age': 'Idade',
    'Ticket_Price': 'Preco_do_Ingresso',
    'Movie_Genre': 'Genero',
    'Seat_Type': 'Tipo_de_Assento',
    'Number_of_Person': 'Numero_de_Pessoas',
    'Purchase_Again': 'Compraria_Novamente'
}, inplace = True)

df.head(1)
```

```python
#8-Exibir os dados das linhas 35, 38, 42 e 50.
df.loc[[35, 38, 42, 50]]
```

```python
#9-Exibir os dados em que o tipo de assento é Standard.
df[df["Tipo_de_Assento"] == "Standard"]
```

```python
#10-Exibir os dados em que as idades dos consumidores está entre 20 (inclusive) e 45 anos (inclusive).
df.loc[(df['Idade'] > 20) & (df['Idade'] < 45)]
```

```python
#11-Criar um novo data frame em que o gênero dos filmes assistidos é Comedy.
df2 = df.loc[df['Genero'] == 'Comedy']
print(df2)
```

```python
#12-Criar um novo data frame exibindo os registros de Preço do Ingresso, Gênero e Tipo de Assento nos clientes que comprariam novamente.
df3= df[["Preco_do_Ingresso", "Genero", "Tipo_de_Assento"]].loc[df["Compraria_Novamente"] == "Yes"]
print(df3)
```
